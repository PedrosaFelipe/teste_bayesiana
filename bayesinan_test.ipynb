```python
import pandas as pd
```

```python
df = pd.read_csv('data_experiment.csv')
```

```python
df
```

```python
df = df.reset_index().rename(columns={'index' : 'day'} )


df = df.pivot( index = 'day' , columns = 'group' , values=['click' , 'visit']).fillna (0)
df.columns = ['click_control' , 'click_treatment' , 'visit_control' , 'visit_treatment']

df = df[['visit_control' , 'click_control' , 'visit_treatment' , 'click_treatment']]
df

df['acc_visit_A'] = df['visit_control'].cumsum()
df['acc_click_A'] = df['click_control'].cumsum()

df['acc_visit_A'] = df['visit_treatment'].cumsum()
df['acc_click_B'] = df['click_treatment'].cumsum()
```
